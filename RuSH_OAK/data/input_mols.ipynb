```python
import rdkit
from rdkit import Chem
from rdkit.Chem import AllChem, Draw

pim447 = Chem.MolFromSmiles("O=C(c1c([H])c([H])c(F)[c@@]([c@@]2c(F)c([H])c([H])c([H])c2F)n1)N(c3c([H])nc([H])c([H])c3[C@@]4([H])C([H])([H])[C@](N([H])[H])([H])C([H])([H])[C@@](C4([H])[H])([H])C([H])([H])[H])[H]")
aha006 = Chem.MolFromSmiles("O=S1(N(Cc2ccccc2)[C@H](COc3ccccc3)[C@H](O)[C@@H](O)[C@@H](COc4ccccc4)N1Cc5ccccc5)=O")
sr3737 = Chem.MolFromSmiles("COc1c(OC)c(OC)cc(NC(c2cccc(n3ncc4c3ccc(Nc5c(F)cccc5)c4)c2)=O)c1")
lre001 = Chem.MolFromSmiles("Nc1nc(N(C2CC2)Cc3sccc3)cc(Cl)n1")

pim447_3d = Chem.MolFromMolFile("PDB_structures/pim447.sdf")
aha006_3d = Chem.MolFromMolFile("PDB_structures/aha006.sdf")
sr3737_3d = Chem.MolFromMolFile("PDB_structures/sr3737.sdf")
lre001_3d = Chem.MolFromMolFile("PDB_structures/lre001.sdf")

pim447_linker = Chem.MolFromSmiles("O=C(N[*])c1nc([*])c(F)cc1")
aha006_linker = Chem.MolFromSmiles("O=S1(N([C@@H]([C@@H]([C@H]([C@H](N1C[*])C[*])O)O)C[*])C[*])=O")
sr3737_linker = Chem.MolFromSmiles("[*]n1c2ccc([*])cc2cn1")
lre001_linker = Chem.MolFromSmiles("[*]N(C[*])C1CC1")


pim447_fragments = [
    Chem.MolFromSmiles("C(*)1=C(C2C[C@H](C)C[C@H](N)C2)C=CN=C1"),
    Chem.MolFromSmiles("Fc1c([H])c([H])c([H])c(F)c(*)1"),
]
aha006_fragments = [
    Chem.MolFromSmiles("C1=CC=C(O*)C=C1"),
    Chem.MolFromSmiles("C1=CC=C(O*)C=C1"),
    Chem.MolFromSmiles("C1=CC=C(*)C=C1"),
    Chem.MolFromSmiles("C1=CC=C(*)C=C1"),
]
sr3737_fragments = [
    Chem.MolFromSmiles("COC1=CC(NC(C2=CC=CC(*)=C2)=O)=CC(OC)=C1OC"),
    Chem.MolFromSmiles("N(*)C1=C(F)C=CC=C1"),
]
lre001_fragments = [
    Chem.MolFromSmiles("Nc1nc(*)cc(Cl)n1"),
    Chem.MolFromSmiles("c(*)1sccc1"),
]
```

```python
Draw.MolsToGridImage([pim447, pim447_3d, pim447_linker, pim447_fragments[0], pim447_fragments[1]], molsPerRow=5, subImgSize=(400,400))
```
